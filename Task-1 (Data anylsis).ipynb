```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
columns = ['number of adults', 'number of children', 'number of weekend nights', 
           'number of week nights', 'car parking space', 'lead time', 'repeated', 
           'P-C', 'P-not-C', 'special requests', 'average price ']

for column in columns:
    df = remove_outliers_iqr(df, column)
```

```python
df.describe()
```

```python
# Histograms for numerical columns
df.hist(bins=30, figsize=(15, 10))
plt.show()

# Count plots for categorical columns
for col in df.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=col, data=df)
    plt.show()
```

```python
# Scatter plots for numerical columns
numerical_cols = df.select_dtypes(include=[np.number]).columns
target_col = 'average price '  # Assuming this is the correct column name

num_cols = len(numerical_cols)
cols_per_row = 3  # Number of columns per row
rows = (num_cols - 1) // cols_per_row + 1  # Calculate number of rows needed

plt.figure(figsize=(16, 12))
for i, col in enumerate(numerical_cols):
    if col != target_col:
        plt.subplot(rows, cols_per_row, i + 1)  # Adjust the grid size dynamically
        sns.scatterplot(x=col, y=target_col, data=df)
        plt.title(f'Scatter plot of {col} vs {target_col}')
plt.tight_layout()
plt.show()
```

```python
# Box plots for categorical columns with fewer unique values
categorical_cols = df.select_dtypes(include=['object']).columns
target_col = 'average price '  # Assuming this is the correct column name
unique_threshold = 20  # Maximum number of unique values to consider

num_cols = sum(df[col].nunique() < unique_threshold for col in categorical_cols)
cols_per_row = 3  # Number of columns per row
rows = (num_cols - 1) // cols_per_row + 1  # Calculate number of rows needed

plt.figure(figsize=(16, 12))
plot_index = 1
for col in categorical_cols:
    if df[col].nunique() < unique_threshold:
        plt.subplot(rows, cols_per_row, plot_index)
        sns.boxplot(x=col, y=target_col, data=df)
        plt.title(f'Box plot of {col} vs {target_col}')
        plot_index += 1
plt.tight_layout()
plt.show()
```

```python
# Exclude non-numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
